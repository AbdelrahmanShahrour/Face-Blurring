```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the image
img = cv2.imread('image.png')
```

```python
#Show the image with matplotlib
plt.imshow(img)
plt.show()
```

```python
# Create a face detector object
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
# Blur the faces
for (x, y, w, h) in faces:
    # Create a copy of the original image
    blurred_img = img.copy()
    # Apply Gaussian blur with kernel size 25x25
    blur = cv2.GaussianBlur(blurred_img[y:y+h, x:x+w], (25, 25), 0)
    blurred_img[y:y+h, x:x+w] = blur
    # Replace the corresponding region in the original image with the blurred copy
    img[y:y+h, x:x+w] = blurred_img[y:y+h, x:x+w]
```

```python
# Save the resulting image with blurred faces
cv2.imwrite('blurred_image.jpg', img)
```

```python
#Show the image with matplotlib
plt.imshow(img)
plt.show()
```
